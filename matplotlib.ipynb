># Matplotlib Tutorial

### **:) What is matplotlib ?**
*Matplotlib* is a popular data visualization library in Python. It's often used for creating static, interactive, and animated visualizations in Python. Matplotlib allows you to generate plots, histograms, bar charts, scatter plots, etc., with just a few lines of code

#### **Key Uses of Matplotlib:** 

**Data Visualization:** Simplifies understanding of data by providing visual representations.

**Plotting:** Supports various plots such as line, scatter, bar, histogram, pie, and 3D plots.

**Integration:** Works well with other Python libraries like NumPy and Pandas.

#### **Key Features of Matplotlib:**

**Customizable Plots:** Allows extensive customization of plots, including colors, labels, and styles.

**Subplotting:** Enables multiple plots in a single figure for comparative analysis.

**Interactive and Animated Plots:** Supports dynamic visualizations for real-time data exploration.

**3D Plotting and Exporting:** Offers 3D plotting capabilities and exporting in multiple formats.

Matplotlib is essential for creating insightful visualizations in data science and machine learning.

```python
# A simple line chart

ls1 = [10,20,30,40,50] # x- axis
ls2 = [25,10,20,5,15] # y- axis
plt.plot(ls1,ls2,marker ="*",ms =8)  # ms => 'marker size'

plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.show()
```

```python
# scatter plot representataion

ls1 = [10,20,30,40,50] # x- axis
ls2 = [25,10,20,5,15] # y -axis
plt.scatter(ls1,ls2,marker='*',s=250)

plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("scatter plot")
plt.show()
```

> ## Bar Plot in Matplotlib

- *A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent.* 

- *The bar plots can be plotted horizontally or vertically.*

- *A bar chart describes the comparisons between the discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the numerical value*.

```python
# Countries and there population bar-chart

Country_name = ['India','America','china','japan','Russia']
Country_population = [17.8,4.23,17.39,3.5,14.8]

plt.bar(Country_name,Country_population, color='green',ec ='k' , ls='dashed')

plt.xlabel("contries")
plt.ylabel("Population")
plt.title("bar plot/chart" )
plt.show()
```

> ## **Creating Pie Charts**
With Pyplot, you can use the **pie()** function to draw pie charts:

```python
# Countries and there population pie-chart

plt.pie(
    Country_population,
    labels=Country_name
)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,8))

plt.pie(
    Country_population,
    labels=Country_name,
    autopct='%0.1f',     # autopct => auto percentage caluclating function ,works as '%(it decides the number alingment).(how many decimal we wants)f ' 
    wedgeprops={'ec':"k",'ls':"dashed",'lw':1}
)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,8))

myexplode = [0.3, 0, 0, 0, 0]
plt.pie(
    Country_population,
    labels=Country_name,
    autopct='%0.1f',     # autopct => auto percentage caluclating function ,works as '%(it decides the number alingment).(how many decimal we wants)f ' 
    explode=myexplode,    # used to highlite/apply some distance from the main chart 
    wedgeprops={'ec':"k",'ls':"dashed",'lw':1}
)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,8))

myexplode = [0.3, 0, 0, 0, 0]
plt.pie(
    Country_population,
    labels=Country_name,
    autopct='%0.1f',     # autopct => auto percentage caluclating function ,works as '%(it decides the number alingment).(how many decimal we wants)f ' 
    explode=myexplode,    # used to highlite/apply some distance from the main chart 
    shadow=True,          # Used for shadowing the graph
    wedgeprops={'ec':"k",'ls':"dashed",'lw':1}
)

plt.legend()
plt.show()
```

> ### **Histogram**

**A histogram is used to represent data provided in the form of some groups.** It is an accurate method for the graphical representation of numerical data distribution. It is a type of bar plot where the X-axis represents the bin ranges while the Y-axis gives information about frequency.

 - *tells us how our data is 'distributed'*
 - *Histohgram are use to identify patterns , trends and outliers of data*

```python
age = np.random.randint(1,100,1000)
age
```

```python
bucket = [0,20,25,30,35,40] # range on x-axis (we can decide it manually also...)
plt.hist(age ,bins = bucket , ec ='k')
plt.show()
```

> ### ***BOX PLOT*** 
*A box and whisker plot or diagram (otherwise known as a boxplot), is a graph summarising a set of data. The shape of the boxplot shows how the data is distributed and it also shows any outliers. It is a useful way to compare different sets of data as you can draw more than one boxplot per graph.*

It representation of data using a **five-number summary (minimum, first quartile, median, third quartile, and maximum)**. It is commonly used to visualize the distribution of data and identify outliers.

> Here are some key points about box plots:

        1. The box represents the interquartile range (IOR), which is the range between the first quartile (25th percentile) and the third quartile (75th percentile).

        2. The vertical lines extending from the box represent the minimum and maximum values, excluding outliers.

        3. The median (second quartile) is represented by a horizontal line within the box.

        4.Outliers are represented by individual points outside of the whiskers, which extend from the ends of the box.

        5.Box plots can be used to compare the distribution of data between different groups or categories.

        6.Box plots can be customized by changing the color, style, and other visual properties.

```python
plt.boxplot(age)
plt.show()
```
