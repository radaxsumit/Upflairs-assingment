># Seaborn :)

      Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
    as per syntax :-
                    sns._(any chart name u wanna plot)_(x =' x-axis cordinate' , y =' y-axis cordinate' , data = _anydata_)

```python
### Let import the libraries

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
print("Libraries imported sucessfully :)")
```

```python
## Load an example dataset from the online repository (requires internet).
# This function provides quick access to a small number of example datasets that are useful for documenting seaborn or generating reproducible examples 
# Note :- that some of the datasets have a small amount of preprocessing applied to define a proper ordering for categorical variables.


df = sns.load_dataset('tips')
df.head()
```

```python
# Fetching all the necessary info about the data
df.info()
```

```python
# plt.scatter() by using matplot or. 
sns.scatterplot(x='tip' , y='total_bill' , data=df , color='#FFBF00')
plt.show( )
```

```python
# sns.lineplot()
sns.lineplot(x='tip', y='total_bill' ,data=df,hue='sex')
plt.show( )
```

```python
# sns.histplot()
sns.histplot(x='tip', y='total_bill' ,data=df)
plt.show()
```

```python
# sns.barplot()
sns.barplot(x='tip', y='total_bill' ,data=df)
plt.show()
```

```python
# sns.boxenplot()
sns.boxenplot(x='tip', y='total_bill' ,data=df)
plt.show()
```

```python
# df['day'].value_counts()

sns.countplot(df['sex'] , color ='#CCCCFF')
plt.show()
```

```python
sns.countplot(df['day'] , color ='#1abc9c')
plt.show()
```

```python
## There are two functions in 'Seaborn' by the help of we can create all the necessary charts of our data.
# 1. relplot => relatational function (both colums should be Numerical)
# 1. catplot => Catagorical function (1 colums should be Numerical and another one should be catogrical)
```

```python
# sns.relplot(x='tip', y='total_bill' , data = df )
sns.relplot(x='tip', y='total_bill' , data = df, hue='time', style='smoker',col='day')
plt.show()
```

```python
# Passing different paramater

sns.relplot(x='tip', y='total_bill' , data = df, hue='sex', style='smoker',col='day')
plt.show()
```

```python
# .corr() => Co-relation function use to tell the total relation

df[['total_bill','tip','size']].corr()
```

```python
df.head(2)
```

```python
## catplot => 1 numerical , 1 catogrial

sns.catplot(x='day',y='tip',data =df,hue='sex')
plt.show()
```

```python
# Kind (parameter) = 'box','bar','strip','boxen','point','violin'   , "count" -> only take one axis

sns.catplot(x='day',y='tip',data =df,hue='sex',kind='box')
plt.show()
```

```python
sns.catplot(x='day',y='total_bill',data =df,hue='sex',kind='bar')
plt.show()
```

```python
sns.catplot(x='time',y= 'tip',data =df,hue='sex',kind='violin')
plt.show()
```

```python
sns.catplot(x='smoker',y='tip',data =df,hue='sex',kind='point')
plt.show()
```

```python
sns.histplot(df['total_bill'],color='#2ecc71')
plt.title("Histogram with kde(kernal density estimate :-)")
plt.show()
```

```python
sns.histplot(df['total_bill'],color='#2ecc71',kde=True)
plt.title("Histogram with kde(kernal density estimate :-)")
plt.show()
```

```python
sns.kdeplot(df['tip'])
plt.show()
```

```python
sns.kdeplot(df['size'])
```
